# Parallel-Coordinates Plot

```python
data = {
    "Id":[1,2,3,4,5,6,7,8,9],
    "Age":[50,25,30,65,55,75,45,70,40],
    "Weight":[160,130,200,150,170,165,180,190,200],
    "Height":[5,6,4,4.5,5.2,4.8,5.8,4.8,5.8]
}

df = pd.DataFrame(data)

#plot
plt.figure(figsize = (10,6))
pd.plotting.parallel_coordinates(df, class_column = 'Id', cols=['Age','Weight','Height'])

plt.title('Parallel Coordinates Plot')
plt.xlabel('Attributes')
plt.ylabel('Values')

plt.show()
```

## Analysis of Outcome

There seems to be an even distribution of ages and heights. There are more individuals that have higher weights. 

There is no clear relationship between an individual's weight and height based on the plot. There are people who are heights but higher weights like the 30-year-old. There are people with larger heights and smaller weights like the 25 year old. This also showcases the lack of relationship between age and the attributes height and weight. The youngest two individuals, age 25 and 30, weigh both the least and the most and are the tallest and the shortest, respectively.
