## Imports

```python
import numpy as np
import logging
import matplotlib.pyplot as plt
import statsmodels.api as sm


import scipy.stats as stats
logger = logging.getLogger(__name__)
```

## Prepare Data

```python
logger.info('Generating datasets...')
normalDataset = np.random.normal(loc=10.0, scale=2.0, size=(1000))
                                # where loc=mean and scale=variance
uniformDataset = np.random.uniform(low=4.0, high=16.0, size=(1000))
logger.info('Datasets created!')

# ** Reference(s)
# * For ".random.normal": https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html#numpy-random-normal
# * For ".random.uniform": https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html#numpy.random.uniform
# **
```

# Q-Q Plot

```python
logger.info('Generating Q-Q Plot...')

# Create Q-Q plot

plt.figure(figsize=(10, 6))
stats.probplot(uniformDataset, dist="norm", plot=plt)
plt.title('Q-Q Plot: Uniform vs. Normal Distribution')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')

# Add quantile markers at 10%, 25%, 50%, 75%
quantiles = [0.1, 0.25, 0.5, 0.75]
for q in quantiles:
    quantileValue = np.quantile(uniformDataset, q)
    theoreticalQuantile = stats.norm.ppf(q, loc=10, scale=2)
    plt.plot([theoreticalQuantile], [quantileValue])
    plt.text(theoreticalQuantile, quantileValue, f'{int(q*100)}%', fontsize=10, verticalalignment='bottom')

plt.grid()
plt.axline((0, 0), slope=1, color='gray', linestyle='--')  # Reference line
plt.show()

# ** Reference(s)
# * https://www.geeksforgeeks.org/quantile-quantile-plots/
# * https://www.statology.org/q-q-plot-python/
# * https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.probplot.html
# **
```
